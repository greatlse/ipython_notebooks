# Simple PCA on images: reducing color "complexity" as a visual example of PCA

This notebook decomposes RGB images into their principal components.  This is done mainly as an exercise for intuitive, visual demonstration of PCA.  In theory, this could have some application in compression of images.

## Importing required libraries
This notebook uses the scikit module `sklearn` as the engine for PCA.  This requires `scipy` and `numpy`.  A few other libraries are required for easy image importing.  `matplotlib` is used for displaying images.

```python
#PCA on the baboon image
%matplotlib inline
import sys
import os
import numpy as np
from scipy import ndimage as ndi
from scipy import misc
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn import decomposition
from skimage import img_as_float
from skimage import img_as_ubyte
```

## An example RGB image and its representation in python

```python
#Show the starting image
baboonfile = '/Users/curt/Downloads/BaboonRGB.jpg'
im = misc.imread(baboonfile)
plt.imshow(im)
```

```python
#verify the image as read by scipy.misc is a 512x512x3 array of np.uint8s between 0 and 255
print im.shape
print type(im)
print type(im[100,203,2])
print np.max(im), np.min(im)
```

## Prepping for PCA: data types and data reshaping
PCA algorithms are designed to work on a matrix (2D) of data.  Rows of the matrix are "samples", which in this case will be pixels from the image.  Columns are the multivariate data "responses" obtained at each "sample".  In this case the columns are three: red, green, and blue.  The matrix elements are the intensities of the corresponding color of a particular pixel.  For this application, the spatial arrangement of pixels is irrelevant, and we want to reshape the data from a 3D matrix with $x$, $y$, and $clr$ values to a flat matrix with only $row$ and $clr$ values.  The example image is 512 pixels in both height and width, so there are $512^2=262144$ pixels in the image. Reshaping is easy to do in `numpy`.  After PCA we will unflatten the data back to a 3D array for displaying results.

```python
#flatten the image in x & y dimensions to prep for pca
imflat = np.array([np.array(im[:,:,i]).flatten() for i in range(im.shape[2])]).T
print imflat.shape
print type(imflat)
print type(imflat[1,1])
```

## Doing PCA via singular value decomposition
The `sklearn` module `decomposition` includes routines for PCA decomposition of our reshaped 262,144 by 3 matrix via singular-value decomposition.  The summary statistics printed are the percent of the variance (and below it, the cumulative percent of the variance) explained by each principal component.  The results show that 65% of the variance is "explained" by the first prinicipal component, and an additional 31% is "explained" by PC2.  Thus, this three-color image could be represented as a two-color image and only 3.9% of the original information would be lost.

```python
#do pca via SVD method
pca = decomposition.PCA(n_components=3)
pca.fit(imflat.astype(float))
imnewflat = pca.transform(imflat)
print pca.explained_variance_ratio_
print np.cumsum(pca.explained_variance_ratio_)
print pca.components_
```

## Un-reshaping the result and displaying data

```python
#reshape the new images and show the components
imnew = imnewflat.reshape(im.shape)
f, axarr = plt.subplots(2, 3, figsize = (24,12))
axarr[0, 0].imshow(imnew[:,:,0], cmap = cm.Greys_r)
axarr[0, 0].set_title('PCA dimension 1')
axarr[0, 1].imshow(imnew[:,:,1], cmap = cm.Greys_r)
axarr[0, 1].set_title('PCA dimension 2')
axarr[0, 2].imshow(imnew[:,:,2], cmap = cm.Greys_r)
axarr[0, 2].set_title('PCA dimension 3')
axarr[1, 0].imshow(im[:,:,0], cmap = cm.Greys_r)
axarr[1, 0].set_title('original image red channel')
axarr[1, 1].imshow(im[:,:,1], cmap = cm.Greys_r)
axarr[1, 1].set_title('original image green channel')
axarr[1, 2].imshow(im[:,:,2], cmap = cm.Greys_r)
axarr[1, 2].set_title('original image blue channel')
```

The charts above display two independent rows of information.  The first row is the three PCA images.  The upper left panel is the image of "scores" in the PC1 dimension, and the upper middle is the score image in the PC2 dimension, etc.  Close inspection reveals that the image contrast varies well with the percent variance explained by each PC.  For example, the upper right panel appears low quality, with few brightly contrasting pixels, which is consistent with PC3 explaining only 4% of the variance in pixel hue in this particular image.  The second image mainly seems to contrast the central nose stripe and eyes from everything else.  In contrast the first image has bright contrast in nearly all regions of the image.

### Color display of PCA results

We can also meld the PCA dimensions back into RGB space one by one for color comparison of how well each adding each dimension successively improves our representation of the original image.  I do this somewhat manually in the code below.  As expected, including all three PC dimensions (100% of the variance) recapitulates the original image.  Also, in nice agreement with the cumulative 96% of the variance explained by only PCs 1 and 2, the middle image is a very reasonable facsimile of the original.  The left image is the "best" 1D i.e. grayscale representation available. 

Inspection of the code will reveal that I had to mask certain blue intensity values in the `pc1` image blue channel.  I suspect this is due to numerical errors from coercing the PCA results back into the variable type `uint8` for preservation of the RGB values. 

```python
pc123 = pca.inverse_transform(imnewflat).astype('uint8')

imnf12 = imnewflat.copy()
imnf12[:, 2] = 0
pc12 = pca.inverse_transform(imnf12).astype('uint8')

imnf1 = imnewflat.copy()
imnf1[:, 2] = 0
imnf1[:, 1] = 0
pc1 = pca.inverse_transform(imnf1).astype('uint8')
pc1[pc1[:, 2] >= 238] = 0

f, axarr = plt.subplots(1, 3, figsize = (24,12))
axarr[0].imshow(pc1.reshape(im.shape))
axarr[0].set_title('PC1 mapped back into RGB space')
axarr[1].imshow(pc12.reshape(im.shape))
axarr[1].set_title('PCs 1 & 2 in RGB space')
axarr[2].imshow(pc123.reshape(im.shape))
axarr[2].set_title('All PCs in RGB space')
```

## Traditional PCA plots
Traditionally, PCA results are illustrated with (i) a scatter plot of the "samples" in the first two (or three) PC dimensions, and (ii) a loadings plot showing how the original data variables (here: colors) map onto the new PC variables.  We have too many pixels to show them all on the scatter plot, but below I choose a random sample of 5000 pixels to plot.

```python
scores = pca.score_samples(imflat)
from numpy import random
randpixels = random.choice(imflat.shape[0], size=5000)
randscores = imflat[randpixels, :]
print pca.components_
```

```python
f, ax = plt.subplots(1, 2, figsize=(12,6))
ax[0].scatter(randscores[:,0], randscores[:,1], color=img_as_float(imflat[randpixels]))
ax[0].set_xlabel('PC1')
ax[0].set_ylabel('PC2')
ax[1].scatter(pca.components_[0, :], pca.components_[1, :], color=[[1,0,0],[0,1,0],[0,0,1]], s=40)
ax[1].set_xlabel('loadings on PC1')
ax[1].set_ylabel('loadings on PC2')
```

The pixel scatter plot on the left colors pixels by their original RGB color.  It shows that PC1 mainly separates black/dark blue pixels from intensely red pixels.  PC2 separates light blue pixels from the rest.  The regions of the graph where you see differently covered points overlapping, such as in the central band where blue-gray pixels overlap brown-orange ones, are spots where PC3, the only remaining dimension in this 3D dataset, must be important.
