```python
#PCA on the baboon image (only 3 RGB channels instead of 500,000 m/z channels like for MSI)
%matplotlib inline
import sys
import os
import numpy as np
from scipy import ndimage as ndi
from scipy import misc
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn import decomposition
```

```python
#Show the starting image
baboonfile = '/Users/curt/Downloads/BaboonRGB.jpg'
im = misc.imread(baboonfile)
plt.imshow(im)
```

```python
#verify the image as read by scipy.misc is a 512x512x3 array of np.uint8s between 0 and 255
print im.shape
print type(im)
print type(im[100,203,2])
print np.max(im), np.min(im)
```

```python
#flatten the image in x & y dimensions to prep for pca
imflat = np.array([np.array(im[:,:,i]).flatten() for i in range(im.shape[2])]).T
print imflat.shape
print type(imflat)
print type(imflat[1,1])
```

```python
#do pca via SVD method
pca = decomposition.PCA(n_components=3)
pca.fit(imflat)
imnewflat = pca.transform(imflat)
pca.explained_variance_ratio_
```

```python
#reshape the new images and show the components
imnew = imnewflat.reshape(im.shape)
f, axarr = plt.subplots(2, 3, figsize = (24,12))
axarr[0, 0].imshow(imnew[:,:,0], cmap = cm.Greys_r)
axarr[0, 0].set_title('PCA dimension 1')
axarr[0, 1].imshow(imnew[:,:,1], cmap = cm.Greys_r)
axarr[0, 1].set_title('PCA dimension 2')
axarr[0, 2].imshow(imnew[:,:,2], cmap = cm.Greys_r)
axarr[0, 2].set_title('PCA dimension 3')
axarr[1, 0].imshow(im[:,:,0], cmap = cm.Greys_r)
axarr[1, 0].set_title('original image red channel')
axarr[1, 1].imshow(im[:,:,1], cmap = cm.Greys_r)
axarr[1, 1].set_title('original image green channel')
axarr[1, 2].imshow(im[:,:,2], cmap = cm.Greys_r)
axarr[1, 2].set_title('original image blue channel')
```
