```python
#k-means on the baboon image
%matplotlib inline
import sys
import os
import numpy as np
from scipy import ndimage as ndi
from scipy import misc
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn import cluster
```

```python
#Show the starting image
baboonfile = '/Users/curt/Downloads/BaboonRGB.jpg'
im = misc.imread(baboonfile)
plt.imshow(im)
```

```python
#verify the image as read by scipy.misc is a 512x512x3 array of np.uint8s between 0 and 255
print im.shape
print type(im)
print type(im[100,203,2])
print np.max(im), np.min(im)
```

```python
#flatten the image in x & y dimensions to prep for k-means
imflat = np.array([np.array(im[:,:,i]).flatten() for i in range(im.shape[2])]).T
print imflat.shape
print type(imflat)
print type(imflat[1,1])
```

```python
#do k-means for k=k
k=4
baboon_kmeans = cluster.KMeans(n_clusters=k, n_jobs=-1)
baboon_kmeans.fit(imflat)
imnewflat = baboon_kmeans.transform(imflat)
```

```python
imclustflat = baboon_kmeans.predict(imflat)
print imnewflat.shape
print imclustflat.shape

nx, ny, ncolor = im.shape
imclust = imclustflat.reshape(nx, ny)
```

```python
#show cluster assignments for each pixel
f, axarr = plt.subplots(1,2, figsize = (12,12))
pic = axarr[1].imshow(imclust, cm.get_cmap('jet', k))
axarr[0].imshow(im)

f.subplots_adjust(right=0.8)
cbar_ax = f.add_axes([0.85, 0.35, 0.02, 0.3])
f.colorbar(pic, cax=cbar_ax, ticks=range(k), label='Cluster index # after k-means with k = %s'%k)
```

```python
#show distances from each pixel to cluster center for all 5 clusters
imnew = imnewflat.reshape(nx, ny, k)

f, axarr = plt.subplots(1,k, figsize = (24,12))
for ind in range(k):
    pic = axarr[ind].imshow(imnew[:, :, ind])
    axarr[ind].set_title('Distance from cluster %s center' % ind)

f.subplots_adjust(right=0.8)
cbar_ax = f.add_axes([0.85, 0.4, 0.05, 0.3])
f.colorbar(pic, cax=cbar_ax)

```

```python
#showing the cluster centers for each cluster (should be k clusters by 3 RGB values)
print baboon_kmeans.cluster_centers_
```

```python
#do k-means for various ks, and tabulate the mean distance as a function of k (elbow method)

klist = range(1,8)
print klist

scores = [];

fits = [];

for k in klist:
    clustering = cluster.KMeans(n_clusters=k, n_jobs=1)
    clustering.fit(imflat)
    fits.append(clustering)
    scores.append(clustering.inertia_)
    print k, clustering.inertia_
```

```python
#k = 3, 4, or 5 seems to be a decent choice; probably 4 is best
plt.scatter(klist, scores)
```
