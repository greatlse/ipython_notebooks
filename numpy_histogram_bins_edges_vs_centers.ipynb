```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dat = np.random.random_integers(low=0, high=2, size = 10)
print dat[0:10]
```

```python
bindat, binedges = np.histogram(a=dat, bins = np.arange(0, 4)-0.5)

print bindat, binedges

print len(bindat)
print len(binedges)
```

```python
#plotting histogram 
plt.figure(figsize=(16, 6))
plt.bar(binedges[:-1], bindat, width=1, facecolor = 'green')
```

```python
#bar plots know that the bin edge is the bin edge, but most other plot methods don't.  They think it is the bin center.
f, ax = plt.subplots(4, 1, figsize=(16, 10))
ax[0].stem(binedges[:-1], bindat)
ax[0].set_xlim([-1, 2.5])
ax[1].plot(binedges[:-1], bindat, 'o-r')
ax[1].set_xlim([-1, 2.5])
ax[2].bar(binedges[:-1], bindat, width=1, facecolor = 'green')
ax[2].set_xlim([-1, 2.5])
ax[3].bar(binedges[:-1], bindat, facecolor = 'gray', width=0.1)
ax[3].set_xlim([-1, 2.5])
```

```python
#we usually would want to use bin centers as the x axis instead of bin edges
bin_centers = binedges[:-1] + np.diff(binedges)/2

print binedges, bin_centers

f, ax = plt.subplots(4, 1, figsize=(16, 10))
ax[0].stem(bin_centers, bindat)
ax[0].set_xlim([-1, 2.5])
ax[1].plot(bin_centers, bindat, 'o-r')
ax[1].set_xlim([-1, 2.5])
ax[2].bar(bin_centers, bindat, width=1, facecolor = 'green')
ax[2].set_xlim([-1, 2.5])
ax[3].bar(bin_centers, bindat, facecolor = 'gray', width=0.1)
ax[3].set_xlim([-1, 2.5])
```
