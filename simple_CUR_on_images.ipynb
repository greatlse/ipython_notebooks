#Doing CUR and plotting displaying results
Here I do CUR on the reshaped RGB image.  Rows of the original data matrix are pixel locations and columns are color channels.  Thus each column of $\mathbf{C}$ will be identical to some column of original the data matrix.  In theory the column chosen will be the best single-color representation of the total three-color image.  

This method is not deterministic and isn't guaranteed to represent your original matrix completely, even in the case where you pick a "low" rank "approximation" that is actually of higher-rank than your original data.

```python
#Reshape data in prep for CUR decomposition
imflat = np.array([np.array(im[:,:,i]).flatten() for i in range(im.shape[2])]).T
print "imflat is a (%s, %s)-shaped matrix: RGB channels are columns and pixels are rows." % imflat.shape

#do CUR decomposition
ncols, nrows = 5, 6
cur = pymf.CUR(imflat, rrank=nrows, crank=ncols)
cur.factorize()

#display results
print "After CUR decomposition, the C matrix is cur.U with shape (%s, %s) and type %s" % (cur.U.shape[0], cur.U.shape[1], cur.U.dtype)
imU = cur.U.reshape(im.shape[0], im.shape[1], nrows).astype('uint8')
print imU.shape
f, ax = plt.subplots(2, ncols, figsize=(20,8))
for rgb in xrange(ncols): 
    ax[0, rgb].imshow(imU[:, :, rgb], cmap=cm.Greys_r)
    ax[0, rgb].set_title('The %s col of C' % rgb)
    if rgb <= 2:
        ax[1, rgb].imshow(im[:, :, rgb], cmap=cm.Greys_r)
        ax[1, rgb].set_title('original image\n%s channel'%(['red', 'green', 'blue'][rgb]))
    else:
        ax[1, rgb].imshow(np.ones_like(im, dtype='uint8')*255, cmap=cm.Greys_r)
        
```

### Visualizing the CUR approximations and comparison to the original image.
Here I vary the row-rank (i.e. number of pixel locations) used in the CUR approximation.  The column-rank is chosen to be one, i.e., only a single color is used in the CUR approximation.

For each varied row-rank, the CUR approximation is found from `cur.U.dot(cur.S.dot(cur.V))`, recast back into a `uint8` data type for compatibility with default color mapping in `matplotlib` (which sometimes necessitates masking when approximated intensities fall out of `uint8` range) and display approximated images (row 0) as well as difference images (row 1).

A very nice feature is that it doesn't take many rows to generate a reasonable approximation to the original.  However a very weird feature is that _increasing the chosen row rank does not gaurantee an increase in the quality of the approximation_.  Repeatedly execute the cell block below to see this.

```python
rranks = [1, 2, 3, 4, 5, 8]
f, ax = plt.subplots(2, ncols=len(rranks)+1, figsize=(20,5))
for ind, rrank in enumerate(rranks):
    
    #do CUR
    cur = pymf.CUR(imflat, rrank=rrank, crank=1)
    cur.factorize()
    Mapprox=cur.U.dot(cur.S.dot(cur.V))
    
    #because RGB display interpretation of matplotlib is dependent on dtype, we need to reconvert \
    #back to uint8 and mask any data outside of the uint8 range
    Mapprox[Mapprox>255] = 255
    Mapprox[Mapprox<0] = 0
    imCUR = Mapprox.reshape(im.shape).astype('uint8')
    
    ax[0, ind].imshow(imCUR)
    ax[0, ind].set_title('rrank=%s'%rrank)
    ax[1, ind].imshow(abs(imCUR.astype('int8')-im.astype('int8')).astype('uint8'))

ax[0, len(rranks)].imshow(im)
ax[0, len(rranks)].set_title("original image")
ax[1, len(rranks)].imshow(im-im)
```

### Showing pixel locations chosen by CUR
Again, the behavior is not deterministic.  White dots indicate chosen pixel location(s) for a row-rank = 1 and column-rank = 1.  (Since some pixels have identical RGB intensities, when a pixel with duplicated/degenerate RGB intensities is chosen by CUR, I plot the location of all of the identical pixels.)

```python
cur = pymf.CUR(imflat, rrank=1, crank=1)
cur.factorize()

diff = abs(imflat-cur.V)==0

locs = np.where(diff.all(axis=1))
locidx = np.unravel_index(locs, dims=(im.shape[0], im.shape[1]))

f, ax = plt.subplots(1, 2, figsize=(20, 10))

ax[0].imshow(im)
ax[0].scatter(locidx[0], locidx[1], s=200, color='white')


Mapprox = cur.U.dot(cur.S.dot(cur.V))
Mapprox[Mapprox>255]=255
Mapprox[Mapprox<0]=0
ax[1].imshow(Mapprox.reshape(im.shape).astype('uint8'))

print 'RGB intensities of the chosen pixel are %s' % (im[locidx[0], locidx[1], :])
print 'location of the IDed pixel(s) is (x=%s, y=%s)' % (locidx[0], locidx[1])
```

## Finding the distribution of approximation quality as a function of row and column rank

I defined "approximation quality" as the ratio of the Frobenius norm of the difference matrix $\mathbf{CUR}-\mathbf{M}$, i.e. if $\mathbf{M}$ is the matrix to be approximated/decomposed, and $\mathbf{CUR}$ is the dot product of the three CUR decomposition matrices, then "approximation quality" is $\frac{\|\mathbf{CUR}-\mathbf{M}\|_F}{\|\mathbf{M}\|_F}$.

```python

def doCURsim(ncol, nrow, crank, rrank):
    zeros = np.zeros(shape=nrow)
    ones  = np.identity(nrow)
    #for a random M
    M = np.random.multivariate_normal(mean=zeros, cov=ones, size=ncol).T
    
    #for a determined M
    #M = np.arange(nrow*ncol).reshape(nrow, ncol)
    Mcur = pymf.CUR(M, rrank=rrank, crank=crank)
    Mcur.factorize()
    diffmat = Mcur.U.dot(Mcur.S.dot(Mcur.V)) - M
    return np.linalg.norm(diffmat) / np.linalg.norm(M)

ndraws = 100
rranks = [3, 4, 5, 6, 7, 8]
cranks = [1, 2, 3, 4, 5, 6]

datshape = (len(rranks), len(cranks), ndraws)
dat = np.empty(datshape)
f, ax = plt.subplots(ncols=len(rranks), nrows=len(cranks), figsize=(20, 30))
vec = 0
for ix, rrank in enumerate(rranks):
    for iy, crank in enumerate(cranks):
        vec = np.asarray([doCURsim(3, 5, crank, rrank) for x in xrange(ndraws)])
        dat[ix, iy, :] = vec
        ax[ix, iy].hist(vec, bins=20)
        ax[ix, iy].set_title('crank=%s\nrrank=%s'%(crank,rrank))
```
